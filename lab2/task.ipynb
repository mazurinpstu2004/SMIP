## Лабораторная работа №2

```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv("../data/lab2/BikeData.csv")
df
```

1. Общее количество арендованных велосипедов за каждый час

```python
df['Total Count'] = df['Partner 1'] + df['Partner 2']
df
```

2. Удаление Partner 1, Partner 2

```python
df = df.drop(columns=['Partner 1', 'Partner 2'])
df
```

3. Преобразование Date к стандартной форме

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
df
```

4. Количество велосипедов выданных в день и месяц, совпадающее текущим числом + день недели.

```python
current_date = datetime.now().date()
current_month = current_date.month
current_day = current_date.day

current_date_df = df[(df['Date'].dt.month == current_month) & (df['Date'].dt.day == current_day)]
current_date_count = current_date_df['Total Count'].sum()
table_day = current_date_df['Date'].dt.weekday.iloc[0]
display(current_date_df)
print(f"Количество:, {current_date_count}, \nДень недели:, {table_day}")
```

5. Functioning Day в boolean, количество нерабочих дней

```python
def get_func_day_state(state):
    if state == 'Yes':
        return True
    else:
        return False

df['Functioning Day'] = df['Functioning Day'].apply(get_func_day_state)
display(df)

functioning_hours = df.groupby(df['Date'])['Functioning Day'].sum()
display(functioning_hours)
not_functioning_days_count = (functioning_hours <= 12).sum()

print(f"Количество нерабочих дней: {not_functioning_days_count}")
```

6. Holiday в 0 или 1

```python
df['Holiday'] = df['Holiday'].apply(lambda x: 1 if x == 'Holiday' else 0)
display(df)
```

7. Temperature Category

```python
def set_temperature_category(temperature):
    if temperature < 0:
        return "Freezing"
    elif (temperature >= 0) & (temperature < 15 ):
        return "Chilly"
    elif (temperature >= 15) & (temperature < 26):
        return "Nice"
    elif (temperature >= 26):
        return "Hot"

df['Temperature Category'] = df['Temperature'].apply(set_temperature_category)
df
```

8. Good weather и ее процент

```python
df['Good weather'] = df.apply(lambda x: 1 if (x['Temperature Category'] == 'Nice') & (x['Humidity'] >= 40) & (x['Humidity'] <= 60) & (x['Wind speed'] < 5.4) & (x['Rainfall'] == 0.0) & (x['Snowfall'] == 0.0) else 0, axis=1)

good_weather_percent = (df[df['Good weather'] == 1].shape[0] / df.shape[0]) * 100
display(df)
print(f"Хорошая погода - {round(good_weather_percent, 2)} %")
```

9. Группировка, сводная таблица

```python
grouped = df.groupby(['Seasons', 'Temperature Category'])['Total Count'].sum()

pivot = df.pivot_table(
    index='Seasons',
    columns='Temperature Category',
    values='Total Count',
    aggfunc='sum'
)

display(grouped)
display(pivot)
```

10. Сезон и температура, где выдается наибольшее и наименьшее количество велосипедов

```python
bikes_max_count = grouped.max()
bikes_min_count = grouped.min()

max_season, max_temp = grouped.idxmax()
min_season, min_temp = grouped.idxmin()

print(f"Наибольшее количество выданных велосипедов - сезон: {max_season}, погода: {max_temp} - {bikes_max_count}")
print(f"Наименьшее количество выданных велосипедов - сезон: {min_season}, погода: {min_temp} - {bikes_min_count}")
```
