```python
import pandas as pd
```

1. Загрузка таблицы html

```python
df = pd.read_html("https://pstu.ru/files/file/Abitur/2025%20final/%D0%A1%D0%A4_%D0%91_%D0%9D_000006978.html")[0]
df
```

2. Удаление лишних элементов из таблицы

```python
cols = df.iloc[12].tolist()
df.columns = cols
df_cleared = df.iloc[13:]
df_cleared = df_cleared.reset_index(drop=True)
df_cleared = df_cleared.dropna(axis=1)
df_cleared = df_cleared.drop(columns=['№', 'Уникальный код', 'Представление приказа', 'Учебная группа'])
display(df_cleared)
```

3. Приведение типов

```python
display(df_cleared.dtypes[df_cleared.columns])
df_numeric = df_cleared[df_cleared.columns].astype(int)
print(50*'-')
display(df_numeric.dtypes[df_numeric.columns])

#df_cleared.info()
#print(70*'-')
#df_numeric.info()
```

4. Средний, минимальный и максимальный суммарные баллы

```python
#avg_points = (df_numeric['Сумма баллов'].sum() / df_numeric['Сумма баллов'].count()).astype(int)

avg_points = (df_numeric['Сумма баллов'].mean()).astype(int)

min_points = df_numeric['Сумма баллов'].min()

max_points = df_numeric['Сумма баллов'].max()

print("Среднее суммарных баллов = ", avg_points,
      "\nМинимальный суммарный балл = ", min_points,
      "\nМаксимальный суммарный балл = ", max_points)

```

5. Среднее по предметам

```python
avg_points_math = (df_numeric['АиНМА / Математика'].mean()).astype(int)
avg_points_cpc = (df_numeric['ПФ / Информатика и ИКТ / Физика / Химия'].mean()).astype(int)
avg_points_rl = (df_numeric['Русский язык'].mean()).astype(int)

print('Таблица среднего балла по предметам')

avg_table = pd.DataFrame()
avg_table['АиНМА / Математика'] = [avg_points_math]
avg_table['ПФ / Информатика и ИКТ / Физика / Химия'] = [avg_points_cpc]
avg_table['Русский язык'] = [avg_points_rl]
avg_table.index = ['Балл']

max_avg_points = avg_table.max()
max_avg_sub = avg_table.idxmax(axis=1).iloc[0]
display(avg_table)
print("Предмет с наибольшим средним баллом: ", max_avg_sub)

```

6. Студенты типа 1

```python
type1_count = df_numeric[(df_numeric['Русский язык'] < avg_points_rl) & (df_numeric['ПФ / Информатика и ИКТ / Физика / Химия'] > avg_points_cpc) & (df_numeric['АиНМА / Математика'] > avg_points_math)].shape[0]
print("Количество студентов типа 1: ", type1_count)
```

7. Студенты типа 2

```python
type2_count = df_numeric[(df_numeric['Русский язык'] > avg_points_rl) & (df_numeric['ПФ / Информатика и ИКТ / Физика / Химия'] < avg_points_cpc) & (df_numeric['АиНМА / Математика'] < avg_points_math)].shape[0]
print("Количество студентов типа 2: ", type2_count)
```

8. Новый DataFrame

```python
final_cols = ["Подразделение", "Уровень подготовки", "Направление подготовки/специальность", "Год", "Количество мест", "Предметы ЕГЭ",
"Средняя сумма баллов", "Min сумма баллов", "Max сумма баллов", "Предмет с высшим средним", "Кол-во студентов тип_1", "Кол-во студентов тип_2"]
final_df = pd.DataFrame(columns=final_cols)
final_df['Средняя сумма баллов'] = [avg_points]
final_df['Min сумма баллов'] = [min_points]
final_df['Max сумма баллов'] = [max_points]
final_df['Предмет с высшим средним'] = [max_avg_sub]
final_df['Кол-во студентов тип_1'] = [type1_count]
final_df['Кол-во студентов тип_2'] = [type2_count]

sub_list = df_numeric.columns.tolist()[1:]
final_df['Предметы ЕГЭ'] = [sub_list]

department = df.iloc[4, 0]
department = department.split(' - ')[1]
final_df['Подразделение'] = [department]

level = df.iloc[5, 0]
level = level.split(' - ')[1]
final_df['Уровень подготовки'] = [level]

specialization = df.iloc[6, 0]
specialization = specialization.split(' - ')[1]
final_df['Направление подготовки/специальность'] = [specialization]

year = df.iloc[2, 0]
year = year[26:30]
final_df['Год'] = [year]

place_count = df.iloc[10, 0]
place_count = place_count[12:15]
final_df['Количество мест'] = [place_count]

final_df
```
