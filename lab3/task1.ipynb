## Лабораторная работа №3. Задание 1

```python
import pandas as pd
```

Таблица рейтингов

```python
ratings_df = pd.read_csv("../data/lab3/ratings.dat", sep="::", names=["UserID", "MovieID", "Rating", "Timestamp"], engine="python")
ratings_df
```

Таблица пользователей

```python
users_df = pd.read_csv("../data/lab3/users.dat", sep="::", names=["UserID", "Gender", "Age", "Occupation", "Zip-code"], engine="python")
users_df
```

Таблица фильмов

```python
movies_df = pd.read_csv("../data/lab3/movies.dat", sep="::", names=["MovieID", "Title", "Genre"], engine="python", encoding="latin-1")
movies_df
```

1. Средние рейтинги фильма в зависимости от пола

```python
var = 42
user_rating_merge_df = pd.merge(users_df, ratings_df, on="UserID")
user_rating_var_df = user_rating_merge_df[user_rating_merge_df["MovieID"] == var]
female_ratings = user_rating_var_df[user_rating_var_df["Gender"] == "F"]['Rating'].mean()
male_ratings = user_rating_var_df[user_rating_var_df["Gender"] == "M"]['Rating'].mean()
print(f"Средний рейтинг для фильма среди женщин - {round(female_ratings, 2)},\nСредний рейтинг для фильма среди мужчин - {round(male_ratings, 2)}")
movies_df[movies_df["MovieID"] == var]
```

2. Фильмы, имеющее не менее 250 оценок. Средний рейтинг в зависимости от пола

```python
general_df = pd.merge(user_rating_merge_df, movies_df, on="MovieID")

most_rated = general_df.groupby("MovieID").size()
most_rated = most_rated[most_rated >= 250]

most_rated_df = general_df[general_df["MovieID"].isin(most_rated.index)]
ratings = round(most_rated_df.groupby(["MovieID", "Gender"])['Rating'].mean(), 2).unstack()
ratings_df = pd.merge(ratings, movies_df[["MovieID", "Title"]], on="MovieID")
ratings_df = ratings_df.rename(columns={"F": "Rating_F", "M": "Rating_M"})
ratings_df = ratings_df.reindex(columns=["MovieID", "Title", "Rating_F", "Rating_M"])

display(most_rated_df)
display(ratings_df)
```

3. Топ 5 лучших фильмов у женщин и худших у мужчин. Топ 5 фильмов с самой большой разницей у мужчин и женщин.

```python
top_five_F = ratings_df.sort_values(by="Rating_F", ascending=False)[:5]
display(top_five_F)

top_five_M = ratings_df.sort_values(by="Rating_M", ascending=True)[:5]
display(top_five_M)

ratings_df["Difference"] = ratings_df.apply(lambda x: abs(x["Rating_F"] - x["Rating_M"]), axis=1)
top_five_dif = ratings_df.sort_values(by="Difference", ascending=False)[:5]
display(top_five_dif)
```

4. Топ 5 фильмов с наибольшим разногласием

```python
rating_std = general_df.groupby("MovieID")["Rating"].std().round(2)
rating_std.name = "STD"
rating_std_df = pd.merge(rating_std, movies_df[["MovieID", "Title"]], on="MovieID")
rating_std_df = rating_std_df.reindex(columns=["MovieID", "Title", "STD"])

rating_std_df = rating_std_df.sort_values("STD", ascending=False).head(5)
display(rating_std_df)
```

5. Средний рейтинг жанров для различных возрастов. Жанр с наибольшим и наименьшим рейтингами в каждой категории возрастов.

```python
general_df["Genre"] = general_df["Genre"].str.split("|")
exploded_general_df = general_df.explode("Genre")
genre_ratings = round(exploded_general_df.groupby(["Genre", "Age"])["Rating"].mean(), 2).reset_index()
print("Средняя оценка жанров по возрасту")
display(genre_ratings)

max_rating_by_age = genre_ratings.loc[genre_ratings.groupby("Age")["Rating"].idxmax()]
min_rating_by_age = genre_ratings.loc[genre_ratings.groupby("Age")["Rating"].idxmin()]

print("Максимальные оценки по возрастам:")
display(max_rating_by_age)

print("Минимальные оценки по возрастам:")
display(min_rating_by_age)
```
